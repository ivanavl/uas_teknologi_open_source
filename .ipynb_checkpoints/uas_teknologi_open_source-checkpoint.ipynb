# Decision Tree Deaths Caused by Covid

```python
#check null data
covid_df.isnull().sum()
```

```python
covid_df.shape
```

```python
covid_death_df = pd.DataFrame(covid_df.dropna())
covid_death_df.reset_index(inplace=True)
covid_death_df.drop(columns='index',inplace=True)
covid_death_df.isnull().sum()
```

#### Remove "missing" and "unknown" value in hosp_yn, icu_yn, death_yn, medcond_yn

```python
covid_death_df.shape
```

```python
covid_death_df = pd.DataFrame(covid_death_df.loc[(covid_death_df.hosp_yn != 'Missing') & (covid_death_df.hosp_yn != 'Unknown') & (covid_death_df.icu_yn != 'Missing') & (covid_death_df.icu_yn != 'Unknown') & (covid_death_df.death_yn != 'Missing') & (covid_death_df.death_yn != 'Unknown') & (covid_death_df.medcond_yn != 'Missing') & (covid_death_df.medcond_yn != 'Unknown'),['death_yn','hosp_yn','icu_yn','medcond_yn']])
covid_death_df.reset_index(inplace=True)
covid_death_df.drop(columns='index',inplace=True)
covid_death_df.shape
```

```python
covid_death_df['hosp_yn'] = covid_death_df['hosp_yn'].replace(['Yes','No'],['1','0'])
covid_death_df['icu_yn'] = covid_death_df['icu_yn'].replace(['Yes','No'],['1','0'])
covid_death_df['death_yn'] = covid_death_df['death_yn'].replace(['Yes','No'],['1','0'])
covid_death_df['medcond_yn'] = covid_death_df['medcond_yn'].replace(['Yes','No'],['1','0'])
covid_death_df
```

```python
#train dataset

from sklearn.model_selection import train_test_split
X = covid_death_df[['hosp_yn','icu_yn','medcond_yn']]
y = covid_death_df.death_yn

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state = 1, train_size = 0.6)
```

```python
#Make Prediction
from sklearn import tree

dt_clf = tree.DecisionTreeClassifier(random_state=1)
dt_clf.fit(X_train, y_train)

# generating predictions
y_hat = dt_clf.predict(X_test)
```

```python
#Show Tree
feature_names = ['hosp_yn','icu_yn','medcond_yn']
class_names = ["Death No", "Death Yes"]

fig = plt.figure(figsize=(25,20))
plot = tree.plot_tree(dt_clf, 
               feature_names=feature_names,
               class_names=class_names,
               filled=True)
```

```python
#Evaluate Model
# accuracy 
from sklearn.metrics import accuracy_score

accuracy_score(y_test, y_hat)
```
